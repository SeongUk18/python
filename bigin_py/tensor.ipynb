```python
#간단한 수학 문제
키 = 170
신발 = 260

#신발 = 키 * a + b

a= tf.Variable(0.1)
b= tf.Variable(0.2)

opt= tf.keras.optimizers.Adam(learning_rate=0.1)
#경사하강법을 도와주는 친구(keras.optimizers)

#opt.minimize(손실함수, var_list=[a,b])
#var_list => 경사하강법으로 업데이트할 weight Variable 목록
#손실 함수 => 실제값 - 예측값 => 오차

def 손실함수():
    예측값 = 키 * a- b
    return tf.square(260- 예측값)
    #square : 제곱해주는 함수
#opt.minimize(손실함수, var_list=[a,b])
# 경사하강 1번 해줌

for i in range(300):
    opt.minimize(손실함수, var_list=[a,b])
    #반복시킴
    print(a.numpy(), b.numpy())
    
```

```python
#대학원 합격 확률 구하기
import pandas as pd
#엑셀 데이터를 다룰때 사용

data= pd.read_csv('gpascore.csv')

data=data.dropna()
#pandas 사용은 아랫 칸 참조

y데이터 = data['admit'].values
#values -> 해당 값을 리스트로 담아줌
x데이터 = []

for i, rows in data.iterrows():
    #iterrows 함수: data라는 데이터 프레임을 가로 한줄씩 출력
    x데이터.append([rows['gre'], rows['gpa'], rows['rank']])

#일반 리스트를 numpy array로 변환해야 tensorflow에 집어 넣기 가능
import numpy as np
#numpy -> 다차원 리스트, 행렬 만들때 사용

    
import tensorflow as tf
#keras : w값 찾거나 할때 유용한 도구
model= tf.keras.models.Sequential([tf.keras.layers.Dense(64,activation='tanh'),tf.keras.layers.Dense(128,activation='tanh'),tf.keras.layers.Dense(1,activation='sigmoid')]) #딥러닝 모델
#tf.keras.layers.Dense(node 갯수,activation 함수) 기본적인 히든 레이어 만드는 법
#노드 갯수의 경우 숫자 상관없음 but 2에 제곱수로 많이 함, 마지막은 신경써서 해야함(결과) 결과는정수로 잘 안나옴

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
#optimizer -> 경사 하강법할때, 얼마나 빼줄지(러닝메이트) => Adam 많이씀
#loss -> 손실 함수 => binary_crossentropy확률 계산할때 좋음
#metrics -> 어떤 요소로 평가 할지

model.fit(np.array(x데이터), np.array(y데이터), epochs=1000)
#모델 학습시키기
#model.fit(x데이터, y 데이터, epochs=10)
#x 데이터는 학습데이터, y데이터는 실제 정답 epo 학습시킬 횟수


#예측
예측값=model.predict([[750,3.70,3],[400,2.2,1]])
#모델을 통해서 예측해주는 함수
print(예측값)
```
